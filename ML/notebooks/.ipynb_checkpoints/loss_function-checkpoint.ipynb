```python
import numpy as np, matplotlib.pyplot as plt
```

```python
y_true = np.arange(0,10)
y_pred = y_true/2.
y1 = np.abs(y_true-y_pred)
y2 = np.abs((y_true-y_pred)/y_true)
y3 = ((y_true-y_pred)/y_true)
plt.figure(figsize=(15,10))
plt.plot(y_true,'m')
plt.plot(y_true**2,'k')
plt.plot(y_true**3,'r')
plt.plot(y_true**3/2,'b')
plt.show()
```

```python
y_hat = np.linspace(0,1,20,endpoint=True)
y_true = np.linspace(0,1,20)/1.1
```

### Notes

Find the equation for the regression line.
    \begin{equation*}   \hat{Y}_i = \hat{\beta}_0 + \hat{\beta}_i X_i + \hat{\epsilon}_i   \end{equation*}  

N: Number of training examples  
$Y_i$: True Label   
$\hat{Y}_i$: Prediction for ith training example  
$\hat{\beta}_i$: Model weight  
$\hat{\beta}_0$: Bias parameter    
$\hat{\epsilon}_i$: Erorr 

### Root Mean Square Error

Insert X values in the equation in order to get the respective Y values  
  \begin{equation*} \hat{Y}_i \end{equation*}  

Now subtract the new Y values (i.e. $\hat{Y}_i$) from the original Y values. Thus, found values are the error terms. It is also known as the vertical distance of the given point from the regression line.
   \begin{equation*}  \hat{Y}_i - Y_i  \end{equation*} 

Square the errors 
   \begin{equation*}  {(\hat{Y}_i - Y_i)}^2  \end{equation*} 

Sum up all the squares.
   \begin{equation*}  \sum_{i=1}^{N}(\hat{Y}_i - Y_i)^2  \end{equation*} 

Divide the value by the total number of observations
   \begin{equation*}  MSE = \sqrt{\frac{1}{N}\sum_{i=1}^{N}(\hat{Y}_i - Y_i)^2}  \end{equation*} 

```python
def rmse(predictions, targets):
    differences = predictions - targets
    differences_squared = differences ** 2
    mean_of_differences_squared = differences_squared.mean()
    rmse_val = np.sqrt(mean_of_differences_squared)
    return [rmse_val,differences,differences_squared]


print("d is: " + str(["%.8f" % elem for elem in y_hat]))
print("\n")
print("p is: " + str(["%.8f" % elem for elem in y_true]))
print("\n")

rmse_val = rmse(y_hat, y_true)

print("rms error is: " + str(rmse_val))
```

```python
plt.figure(figsize=(15,10))
plt.plot(y_hat,rmse_val[2])
plt.xlabel("Predicted Values")
plt.ylabel("RMSE")
```

### Mean Square Error

Insert X values in the equation in order to get the respective Y values  
  \begin{equation*} \hat{Y}_i \end{equation*}  

Now subtract the new Y values (i.e. $\hat{Y}_i$) from the original Y values. Thus, found values are the error terms. It is also known as the vertical distance of the given point from the regression line.
   \begin{equation*}  \hat{Y}_i - Y_i  \end{equation*} 

Square the errors  
   \begin{equation*}  {(\hat{Y}_i - Y_i)}^2  \end{equation*} 

Sum up all the squares.
   \begin{equation*}  \sum_{i=1}^{N}(\hat{Y}_i - Y_i)^2  \end{equation*} 

Divide the value by the total number of observations
   \begin{equation*}  MSE = \frac{1}{N}\sum_{i=1}^{N}(\hat{Y}_i - Y_i)^2  \end{equation*} 

```python
def mse(predictions, targets):
    differences = predictions - targets
    differences_squared = differences ** 2
    mean_of_differences_squared = differences_squared.mean()
    
    return [mean_of_differences_squared,differences_squared]


print("d is: " + str(["%.8f" % elem for elem in y_hat]))
print("\n")
print("p is: " + str(["%.8f" % elem for elem in y_true]))
print("\n")

mse_val = mse(y_hat, y_true)

print("rms error is: " + str(mse_val))
```

```python
plt.figure(figsize=(15,10))
plt.plot(y_hat,mse_val[1])
plt.xlabel("Predicted Values")
plt.ylabel("MSE")
```

### Mean Absolute Error

Insert X values in the equation in order to get the respective Y values  
  \begin{equation*} \hat{Y}_i \end{equation*}  

Now subtract the new Y values (i.e. $\hat{Y}_i$) from the original Y values. Thus, found values are the error terms. It is also known as the vertical distance of the given point from the regression line.
   \begin{equation*}  \hat{Y}_i - Y_i  \end{equation*} 

Take the absolute value of the errors  
   \begin{equation*}  |{(\hat{Y}_i - Y_i)}|  \end{equation*} 

Sum up all the values.
   \begin{equation*}  \sum_{i=1}^{N}|(\hat{Y}_i - Y_i)|  \end{equation*} 

Divide the value by the total number of observations
   \begin{equation*}  MSE = \frac{1}{N}\sum_{i=1}^{N}|(\hat{Y}_i - Y_i)|  \end{equation*} 

```python
def mae(predictions, targets):
    differences = predictions - targets
    absolute_differences = np.absolute(differences)
    mean_absolute_differences = absolute_differences.mean()
    return [mean_absolute_differences,absolute_differences]

mae_val = mae(y_hat, y_true)

print ("mae error is: " + str(mae_val))
```

```python
plt.figure(figsize=(15,10))
plt.plot(y_hat,mae_val[1])
plt.xlabel("Predicted Values")
plt.ylabel("MAE")
```
