The input data is fed to the input layer and then the **neurons** perform a linear transformation on these inputs using the weights and biases.  

Then after that the activation function introduces some non-linearity to the network. Only nonlinear activation functions allow such networks to compute nontrivial problems using only a small number of nodes. Without the activation function our model is simply a linear regression model.   

\begin{equation*}  output = \sum_{i=1}^{N}(weights*input) + bias \end{equation*} 

**N** = the number of neurons connected to a particular node  
**weights** =  a trainable parameters updated during the backprapogation process  
 - Weights are used to connect the each neurons in one layer to the every neurons in the next layer.  

 - Weight determines the strength of the connection of the neurons.   

 - Weights near zero mean changing this input will not change the output.   
 
 
**bias** = bias node a trainable parameters updated during the backpropagation process  
 - Bias means how far off our predictions are from real values.  

 - Low Bias: Suggests more assumptions about the form of the target function.  

 - High-Bias: Suggests less assumptions about the form of the target function.  

```python
import numpy as np, matplotlib.pyplot as plt
from sympy import Symbol
```

### Binary Step Function

   
$f(x) = 1$, $x >= 0$  
$\qquad$      $= 0$,  $x < 0$  

The linear binary step function can be used as an activation function for any binary classifier and is not useful with multiple classes. 

#### Problem
The biggest problem with this activation function is that the gradiant is always zero since this function is a constant. As a result of the gradiant being zero, during the backpropagation process the weights and biases cannot be updated. 

```python
def binary_step(x):
    if x<0:
        return 0
    else:
        return 1

# Find the gradiant
x_ = Symbol('x')
y_1 = 1
y_0 = 0*x_
yprime = y_0.diff(x_)

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,binary_step(i), "ro", alpha = 0.3, markersize=20)
    plt.plot(-i,binary_step(-i), "ro", alpha = 0.3, markersize=20)
    
    # Plot the gradiant of the function
    plt.plot(i,yprime, "k.", markersize=10)
    plt.plot(-i,yprime, "k.", markersize=10)
    
plt.ylim(-1.5,1.5) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### Linear Function
\begin{equation*} f(x) = ax \end{equation*} 

The next step up from this binary function is the linear activation function. This is essentially the identity

#### Problem
The issue with this activation function is that although the gradiant is different from zero it is still a constant and as a result the weights and biases during the backpropagation step will be independent of the input, x, and will be updated at a constant value. So, again this activation function will not capture any complexities of the data.

```python
def linear_function(x,a=4):
    
    return a*x

# Find the gradiant
a=4
x_ = Symbol('x')
y_ = a*x_
yprime = y_.diff(x_)

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,linear_function(i,a=4), "ro", alpha = 0.3)
    plt.plot(-i,linear_function(-i,a=4), "ro", alpha = 0.3)
    
    # Plot the gradiant of the function
    plt.plot(i,yprime, "k.")
    plt.plot(-i,yprime, "k.")
    
plt.ylim(-1.5,1.5) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### Sigmoid or Logistic Activation Function

\begin{equation*} f(x) = \frac{1}{(1+e^{-x})} \end{equation*} 

- It allows you to reduce extreme or atypical values in the data without eliminating them.  
- It converts independent variables of almost infinite range into simple probabilities between 0 and 1 and is often used in binary classification problems.  
- Most of its output will be very close to the extremes of 0 or 1. The sigmoid activation function transforms all input data between 0 and 1. The function is significant for domain -3 and 3.  

Additionally, the sigmoid function is not symmetric around the origin. So the output of all the neurons will be of the same sign. This can be addressed by scaling the sigmoid function which is exactly what happens in the tanh function.  

This function is a smooth S-shaped function and is continuously differentiable. The derivative of this function comes out to be (1-sigmoid(x)).  

The gradient values are significant for domain -3 and 3 but the graph gets much flatter in other regions. This implies that for values greater than 3 or less than -3 will have very small gradients and as the gradient value approaches zero, the network is not really learning.  

```python
def sigmoid_function(x):
    z = (1/(1 + np.exp(-x)))
    return z

"""
Print the results of the activation function. As you can see, for extreme outputs the activation function
transforms these values into 0 or 1.
"""
print(sigmoid_function(100))
print(sigmoid_function(1.1))
print(sigmoid_function(-100))

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,sigmoid_function(i), "ro", alpha = 0.3, markersize=20)
    plt.plot(-i,sigmoid_function(-i), "ro", alpha = 0.3, markersize=20)
    
    # Plot the gradiant of the function. 
    plt.plot(i,1-sigmoid_function(i), "k.", markersize=10)
    plt.plot(-i,1-sigmoid_function(i), "k.", markersize=10)#should be 1-sigmoid_function(-i)🤔
    
plt.ylim(-0.1,1.5) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### Tanh

To fix the lack of symmetry problem we can use the Tanh function. This function is similar to the sigmoid functioon except that it is symmetric around the origin.  

\begin{equation*} Tanh = \frac{2}{1 + {exp}^{-2*x}} -1 \end{equation*} 

Just like the sigmoid function, the Tanh function is differentiable at every point. It is also steeper and as a result the only significant changes in the weights during backpropagation will be for small output values.


```python
def tanh_function(x):
    z = (2/(1 + np.exp(-2*x))) -1
    return z

"""
Print the results of the activation function. As you can see, for extreme outputs the activation function
transforms these values into 0 or 1.
"""
print(tanh_function(100))
print(tanh_function(1.1))
print(tanh_function(-100))

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,tanh_function(i), "ro", alpha = 0.3, markersize=20)
    plt.plot(-i,tanh_function(-i), "ro", alpha = 0.3, markersize=20)
    
    # Plot the gradiant of the function. Not sure why it has that sharp point??? 🤔
    plt.plot(i,1-(tanh_function(i))**2, "k.", markersize=10)
    plt.plot(-i,1-(tanh_function(-i))**2, "k.", markersize=10)
    
plt.ylim(-1.5,1.5) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### ReLU

Rectified Linear Unit is a nonlinear activation function. The advantage of using the ReLU function is that the neurons will only be deactivated if the output of the transformation is less than 0.

\begin{equation*} f(x)=max(0,x) \end{equation*} 

The gradient of the function to the right of the origin is a constant and to the right it is always zero. This means that during the backpropagation process some of the weights never get updated and during the forward propagation the neurons are dead.

```python
def relu_function(x):
    if x<0:
        return 0
    else:
        return x
    
def relu_function_prime(x):
    if x<0:
        return 0
    else:
        return 1

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,relu_function(i), "ro", alpha = 0.3, markersize=20)
    plt.plot(-i,relu_function(-i), "ro", alpha = 0.3, markersize=20)
    
    # Plot the gradiant of the function.
    plt.plot(i,relu_function_prime(i), "k.", markersize=10)
    plt.plot(-i,relu_function_prime(-i), "k.", markersize=10)
    
plt.ylim(-1.5,10) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### Leaky ReLU

The Leaky ReLU addresses the Relu function problem as the output takes on negative values.

$f(x)= 0.01x, \quad x<0$  
$\qquad    =   x, \quad x>=0$

```python
def leaky_relu_function(x):
    if x<0:
        return 0.01*x
    else:
        return x
    
def leaky_relu_function_prime(x):
    if x<0:
        return 0.01
    else:
        return 1   

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,leaky_relu_function(i), "ro", alpha = 0.3, markersize=20)
    plt.plot(-i,leaky_relu_function(-i), "ro", alpha = 0.3, markersize=20)
    
    # Plot the gradiant of the function.
    plt.plot(i,leaky_relu_function_prime(i), "k.", markersize=10)
    plt.plot(-i,leaky_relu_function_prime(-i), "k.", markersize=10)

plt.hlines(0, -20, 20, colors='k', linestyles='solid', label='', linewidth=2.0, alpha=0.3)
plt.ylim(-0.2,10) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### Parameterised ReLU

This is a more general form of the leaky ReLu. The parameter "a" is a trainable parameter that the network learns. If it is set yo 0.01 then this function simplifies to the leaky ReLu.

$f(x)= a*x, \quad x<0$  
$\qquad    =   x, \quad x>=0$

```python
def parameterised_relu_function(x, a=0.1):
    if x<0:
        return a*x
    else:
        return x
    
def parameterised_relu_function_prime(x, a=0.1):
    if x<0:
        return a
    else:
        return 1

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,parameterised_relu_function(i), "ro", alpha = 0.3, markersize=20)
    plt.plot(-i,parameterised_relu_function(-i), "ro", alpha = 0.3, markersize=20)
    
    # Plot the gradiant of the function.
    plt.plot(i,parameterised_relu_function_prime(i), "k.", markersize=10)
    plt.plot(-i,parameterised_relu_function_prime(-i), "k.", markersize=10)

plt.hlines(0, -20, 20, colors='k', linestyles='solid', label='', linewidth=2.0, alpha=0.3)
plt.ylim(-1.5,10) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### Exponential Linear Unit

Exponential Linear Unit (ELU) is a variant of ReLU in that it modifies the slope of the negative part of the function unlike the leaky relu and parametric ReLU functions.

$f(x)= a*(exp(x)-1), \quad x<0$  
$\qquad    =   x, \quad \quad \quad \quad \quad \quad x>=0$


```python
def elu_function(x, a=1.5):
    if x<0:
        return a*(np.exp(x)-1)
    else:
        return x
    
def elu_function_prime(x, a=1.5):
    if x<0:
        return a*np.exp(x)
    else:
        return 1

plt.figure(figsize=(15,10))

for i in np.arange(0,11):
    plt.plot(i,elu_function(i), "ro", alpha = 0.3, markersize=20)
    plt.plot(-i,elu_function(-i), "ro", alpha = 0.3, markersize=20)
    
    # Plot the gradiant of the function.
    plt.plot(i,elu_function_prime(i), "k.", markersize=10)
    plt.plot(-i,elu_function_prime(-i), "k.", markersize=10)

plt.hlines(0, -20, 20, colors='k', linestyles='solid', label='', linewidth=2.0, alpha=0.3)
plt.ylim(-2,10) #Range
plt.xlim(-11,11) #Domain

plt.show()
```

### Softmax

Softmax function is often described as a combination of multiple sigmoids. It can instead be used for multiclass classification problems. The function returns the probability for a datapoint belonging to each individual class.


\begin{equation*} \sigma(z)_j = \frac{z_j}{ sum_{i=1}^{N}e^{z_i}} \end{equation*} 

While building a network for a multiclass problem, the output layer would have as many neurons as the number of classes in the target. For instance if you have three classes, there would be three neurons in the output layer. Suppose you got the output from the neurons as [1.2 , 0.9 , 0.75]. When you apply the softmax function over these values, you will get the following result – [0.42 ,  0.31, 0.27]. These represent the probability for the data point belonging to each class. Note that the sum of all the values is 1. 

```python
def softmax_function(x):
    z = np.exp(x)
    z_ = z/z.sum()
    return z_

def softmax_function_prime(x):
    # Find the gradiant
    x_ = Symbol('x')
    y = softmax_function(x_)
    
    yprime = y.diff(x_)
    return yprime

print(softmax_function([-0.8, 1.2, 3.1]))

plt.figure(figsize=(15,10))

plt.plot(np.linspace(-4,4,100),softmax_function(np.linspace(-4,4,100)), "r.", alpha = 0.3, markersize=10)

plt.show()
```

#### Notes

- Sigmoid functions and their combinations generally work better in the case of classifiers
- Sigmoids and tanh functions are sometimes avoided due to the vanishing gradient problem
- ReLU function is a general activation function and is used in most cases these days
- If we encounter a case of dead neurons in our networks the leaky ReLU function is the best choice
- Always keep in mind that ReLU function should only be used in the hidden layers
- As a rule of thumb, you can begin with using ReLU function and then move over to other activation functions in case ReLU doesn’t provide with optimum results
